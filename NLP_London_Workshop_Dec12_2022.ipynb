```python
from transformers import pipeline

qa_model = pipeline("question-answering")
```

```python
context = "This contract starts on 13 Jan 2023 and ends on 28 Feb 2023"
question = "When does the contract start"
```

```python
qa_model(context=context, question=question, top_k=3)
```
